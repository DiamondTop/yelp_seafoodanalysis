```python
import psycopg2
import pandas as pd
import numpy as np



# DB_HOST = "ec2-54-160-120-28.compute-1.amazonaws.com"
# DB_NAME = "d74fqpl6qtrn0m"
# DB_USER = "bdlzvislimyeil"
# DB_PASS = "c72c07abb2838abbfeca8fc6685a6663b18d8fa417a29170dc34b033d87b373f"

def connect(param_dic):
    """ Connect to the PostgreSQL database server """
    conn = None
    try:
        # connect to the PostgreSQL server
        print('Connecting to the PostgreSQL database...')
        conn = psycopg2.connect(**param_dic)

    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        sys.exit(1)
    return conn


def single_insert(conn, insert_req):
    """ Execute a single INSERT request """
    cursor = conn.cursor()
    try:
        cursor.execute(insert_req)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    cursor.close()


#params_dic = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASS, host=DB_HOST)

def main():
    
    # Connection params
    param_dic = {
        "host"      : "ec2-54-160-120-28.compute-1.amazonaws.com",
        "database"  : "d74fqpl6qtrn0m",
        "user"      : "bdlzvislimyeil",
        "password"  : "c72c07abb2838abbfeca8fc6685a6663b18d8fa417a29170dc34b033d87b373f"
    }

    # Read the csv file
    df = pd.read_csv(r'C:\Users\leroy\Desktop\yelpproject\yelp_seafoodanalysis\data_for_seafood\seabassyelpmostrecom2.csv')
    
    # Connect to the database 
    conn = connect(param_dic)
    
    #Create table called most_recom1 & now mostrecom2
    cur = conn.cursor()
    cur.execute("CREATE TABLE most_recom2 (id SERIAL PRIMARY KEY, Word VARCHAR, Word_freq VARCHAR);")
    
    # Replace NaN values with 'NULL'
    words = [x for x in df.columns if x != 'Word']
    for word in words:
        df[word] = df[word].apply(lambda x: 'NULL' if np.isnan(x) else x)
        
    # Also drop any special character in the country name
    df['Word'] = df['Word'].astype(str).replace('[^a-zA-Z0-9 ]', '', regex=True)     
    
    
    # Insert the dataframe one row at the time
    # For each country, upload the yearly C02 emissions
    for i in df.index:
        word_freq = df['Word_freq'][i]
        
        # Loop through each year
        for word in words:
            word = df[word][i]
            # Build the insert query
            query = """
            INSERT into most_recom2(word, word_freq) values('%s',%s);
            """ % (word, word_freq)
            # Insert into the database
            single_insert(conn, query)
        
    print("All rows were sucessfully inserted in the emissions table")
    
    # Close the database connection
    conn.close()


if __name__ == "__main__":
    main()
```
